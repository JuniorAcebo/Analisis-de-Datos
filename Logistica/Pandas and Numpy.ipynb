### 2.1.3. Archivo olist_order_items_dataset

* Leemos el archivo

Aquí me parece súper útil groupby. Antes lo veía complicado, pero ahora entiendo que es clave para resumir datos.

```python
items_agg = items.groupby(
    ['order_id']).agg(
        # conteo de producto
        {'order_item_id': 'count',
        # suma de los precios de los artículos
        'price': 'sum'}
                      ).reset_index() 
```

### 2.1.6. olist_orders_dataset

Convertimos varias columnas a formato fecha para hacer análisis de tiempo. Me parece muy útil porque si no convierto a fecha, no podría calcular retrasos, meses o trimestres.

```python
## convierte order_purchase_timestamp
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'], errors='coerce' )

# order_approved_at
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])

# order_delivered_carrier_date
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])

# order_delivered_customer_date
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# order_estimated_delivery_date
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
## Ahora podemos ver como ha cambiado el formato:
orders.info()
```

Esto es clave, me gusta porque facilita agrupar por mes o año para ver tendencias.

```python
# Define una columna con el año en que sucedió la orden
orders['year'] = orders['order_purchase_timestamp'].dt.year

# Define una columna con el mes en que sucedió la orden
orders['month'] = orders['order_purchase_timestamp'].dt.month

# Define una columna con trimestre con el que paso la orden (ej. Q12018)
# https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.to_period.html
orders['quarter'] = orders['order_purchase_timestamp'].dt.to_period('Q')

# Define una columna con mes y año con el que paso la orden (ej. 02-2018)
# Hint: ¿que hace el metodo ...to_period('M')?
orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
```

* delta_days: diferencia en días entre entrega real y estimada.
* delay_status: clasifica en a tiempo, pequeño retraso o gran retraso.
  
Esto es de lo más útil del ejercicio. Entender el retraso en días es algo que cualquier empresa necesita.

```python
# Nota: tenemos que realizar la conversion de
# segundos a días

orders['delta_days'] = (
    orders['order_delivered_customer_date'] -
    orders['order_estimated_delivery_date']
    ).dt.total_seconds()/ 60 / 60 / 24
```

```python
# Define 
orders['delay_status']  = np.where(
    orders['delta_days'] > 3, 'long_delay',
    np.where(orders['delta_days'] <= 0, 'on_time','short_delay')
    )
```

## 3. Procesamiento global

### 3.1. Clientes + geolocalización

Para unir dos fuentes de datos, podemos usar la función `.merge`. En el siguiente ejemplo se unes los datos de los clientes junto sus geolocalizaciones.

El .merge() al principio me costó entender, pero ahora lo veo como un JOIN de SQL. Es demasiado útil.

```python
customers_geolocation = customers.merge(
    unique_geolocations,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)
```

### 3.2. Clientes + geolocalización + nombre del estado donde viven
Ahora repetiremos un proceso análogo pero con los nombres del estado donde viven los customers

```python
# Une los dataframe customers_geolocation y states_abbreviations

customers_geolocation_estado = customers_geolocation.merge(
    states_abbreviations,
    left_on='geolocation_state',   # columna en customers_geolocation
    right_on='abbreviation',       # columna en states_abbreviations
    how='left'
)
```

```python
customers_geolocation_estado
```

### 3.3. Órdenes + total de artículos y precios

```python
# une los dataframe orders y items_agg por order_id
orders_totals = orders.merge(
    items_agg,
    on=['order_id'],   # columna común en ambos DataFrames
    how='left'         # conserva todas las filas de orders
)
```

```python
orders_totals
```

### 3.4 Clientes + geolocalización + nombre del estado donde viven + Órdenes + total de artículos y precios

```python
results = orders_totals.merge(
    customers_geolocation_estado,
    on=['customer_id'],
    how='left'
    )
```
